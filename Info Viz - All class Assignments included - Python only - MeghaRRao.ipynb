## Top 20 Artists in Europe and Asia.

```python
# get the rank of artists by quantity
df_rank = df_EA.groupby(['Artist'])['Quantity'].sum().reset_index()
df_rank['Rank'] = df_rank['Quantity'].rank(method='dense', ascending=False) 
```

```python
# get the quantity by artists and genre
# append the rank of artists
df_rank_genre = df_EA.groupby(['Artist', 'Genre'])['Quantity'].sum().reset_index()
df_rank_genre=pd.merge(df_rank_genre, df_rank[['Artist', 'Rank']], on='Artist', how='left')
df_rank_genre.sort_values(by='Rank', inplace=True)
```

```python
df_rank_genre.head()
```

```python
# reformat for plotting
df_top20 = df_rank_genre.loc[df_rank_genre['Rank'] <= 20,
                             ['Artist', 'Genre', 'Quantity']].pivot(columns='Genre',
                                                                    index='Artist',
                                                                    values='Quantity')

df_top20['total'] = df_top20.sum(axis=1)
df_top20.sort_values(by=['total', 'Artist'], ascending=[True, False], inplace=True)
df_top20.drop('total', axis=1, inplace=True)
```

## Q2 -  Rock music's monthly market share fluctuation in the US (Year 2011 & 2012)

As per Wikipedia, market share is the percentage of a market accounted for by a specific entity. It can be Unit Market share (in terms of sales quantity) or Revenue market share (in terms of sales Revenue). As price is almost constant in this dataset, both the percentages have the same monthly value.

#### Revenue market share

Revenue market share reflects the prices at which goods are sold.

    Revenue market share (%) = 100 * Sales Revenue ($) / Total Market Sales Revenue($)
    
#### Note: In the dataframe, column 'Sales' has the revenue generated (which is simply Quantity * UnitPrice)
    
 Reference: Wikipedia


```python
# Let us use the existing dataframe and filter it further using a newly created year column, genre and country
df_UC['InvDate_year'] = pd.to_datetime(df_UC['InvoiceDate']).dt.strftime("%Y")
df_temp=df_UC[(df_UC['Genre']=='Rock') & (df_UC['Country']=='USA') & (df_UC['InvDate_year'].isin(['2011','2012']))].copy()

# Now let us view the monthly sales revenue using groupby and sum functions
df_rockUS = df_temp.groupby(['Genre','Country','InvDate_year_month'])['Sales'].sum().reset_index()
df_rockUS 
```

```python
# Let us use temporary dataframes to filter and groupby data to find the total sales 
df_temp1=df_UC[(df_UC['Country']=='USA') & (df_UC['InvDate_year'].isin(['2011','2012']))].copy()

total_sales = df_temp1.groupby(['InvDate_year_month'])['Sales'].sum().reset_index()
total_sales.rename(columns={"Sales": "Total_Sales"}, inplace=True)
total_sales
```

By visually examining the two dataframes, it is evident that Total sales has more entries than Rock sales. Entries are absent for certain months for Rock sales. Hence, let us do a left join to merge the relevant columns.

Suprisingly, the data is completely missing for certain months (for example: Feb, 2011). Let us choose to ignore entries that are completely missing.

```python
# Merge the total sales and the intial rock music dataframe, and create a new column for month name and year
df_Rock = pd.merge(total_sales,df_rockUS, on=['InvDate_year_month'], how='left')
df_Rock['Month & year'] = pd.to_datetime(df_Rock['InvDate_year_month']).dt.strftime("%b %y")

# Calculate the sales market share percentage
df_Rock['Sales_MarketShare (%)'] = (df_Rock['Sales']*100/df_Rock['Total_Sales']).round(2)

# Set Nan values to zero as Rock has no entries for Sales (meaning zero sales) for these months
df_Rock['Sales_MarketShare (%)'].fillna(0, inplace=True)   
df_Rock
```

If we are sure that the missing entries for Rock Sales (in certain months with overall total sales) is not due to zero sales, we can use the below code to delete the inaccurate entries for market share (saying 0% market share). 

    df_Rock['Sales_MarketShare (%)'].dropna(inplace=True)
    

As per Stephen Few's design principles, we can choose to ignore the null values. However, in the above case, 
as total sales has been recorded correctly, the market share is indeed 0% from the visual examination of the dataset.

### Plotting method 2 - Factorplot

Let us use factorplot to stack the graphs for each year for easier comparison.

```python
# Let us create new columns for Month and Year variables
df_Rock['Invoice_Month'] = pd.to_datetime(df_Rock['InvDate_year_month']).dt.strftime("%b")
df_Rock['Invoice_Year'] = pd.to_datetime(df_Rock['InvDate_year_month']).dt.strftime("%Y")

# Now, let us segregate data to create separate dataframes for each year
df_2011=df_Rock[(df_Rock['Invoice_Year']=='2011')].copy()
df_2012=df_Rock[(df_Rock['Invoice_Year']=='2012')].copy()

# Plot the timeseries plots for each year using factorplot. Set title, labels and ticks.
plt.figure(figsize=(20, 10))
g1=sns.factorplot(x="InvDate_year_month", y = "Sales_MarketShare (%)", hue = "Invoice_Year",data = df_2011, palette='muted')
plt.title("Rock music monthly market share in 2011 and 2012", fontsize =20); 
plt.xlabel('Invoice Month', size=14); plt.ylabel('Sales Revenue - Market Share (%)', size=14);
g2=sns.factorplot(x="InvDate_year_month", y = "Sales_MarketShare (%)", hue = "Invoice_Year",data = df_2012, palette='muted')
g1.set_xticklabels(df_2011['Invoice_Month']);g2.set_xticklabels(df_2012['Invoice_Month'])
plt.xlabel('Invoice Month', size=14); plt.ylabel('Sales Revenue - Market Share (%)', size=14)
```
