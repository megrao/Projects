```python
def generate_wordcloud(words, mask):
    word_cloud = WordCloud(width = 350, height = 350, background_color='white', stopwords=stopwords, mask=mask,contour_width=2, contour_color='black',colormap="RdYlGn").generate(words)
    plt.figure(figsize=(8,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
df1.head()
```

```python
listings = pd.DataFrame(df1.neighbourhood.value_counts())
listings.reset_index(inplace=True)
listings.head()
```

```python
listings.neighbourhood = listings.neighbourhood.astype('object')
listings.rename(columns={'index':'neighbourhood', 'neighbourhood':'no_of_listings'}, inplace=True)
listings.head()
```

```python
import seaborn as sns; sns.set()
plt.figure(figsize=(10,10))
sns.barplot(x="number_of_reviews", y="neighbourhood", orient="h", data=df1, palette="rocket")
plt.xlabel("Number of Bookings")
plt.tight_layout()
plt.savefig("bookings.png")
```

```python
df= pd.merge(df1, listings, on='neighbourhood')
df
```

```python
df['reviews per listing'] = df['number_of_reviews']/df['no_of_listings']
df.sort_values(by='reviews per listing', ascending=False)
```

```python
#Plot histogram to show the spread and shape of the data

plt.figure(figsize=(15,15))

# Show 4 different binwidths
for i, binwidth in enumerate([2500, 5000, 10000, 25000]):
    
    # Set up the plots
    ax = plt.subplot(2, 2, i + 1)
    
    # Draw the plots
    ax.hist(df['number_of_reviews'], bins = int(86000/binwidth),
             color = 'blue', edgecolor = 'black')
    
    # Title and labels
    ax.set_title('Listings v/s no.of reviews = %d' % binwidth, size = 15)
    ax.set_xlabel('Number Of Reviews', size = 12)
    ax.set_ylabel('Listings', size= 12)

plt.show()
```
