> Types of Visualizations

* $Line plots$

SeriesObject.plot()              # for plotting single column Series object against row index

df['col1'].plot(kind=’line’)     # for plotting single column against row index in a dataframe

df.plot()                        # for plotting all columns against row index

df['col1'].plot( figsize=(16,8), style=':b')                # color=’crimson’

* $Bar plots$

df[‘col1’].plot(kind=’bar’)       # plot single column against row index

df.plot(kind=’bar’)               # plot all columns(side by side) against row index

df.plot(kind=’bar’, stacked=True) # plot all columns(stacked up) against row index

df.plot(x='col1',y='col2',kind='bar')  # col1 against col2

df[['col1','col2']].plot(kind='bar')   # col1 and col2 against row index

df[['col1','col2']].sort_values('col1').plot(kind='bar')    # sort against col1

* $Histogram  plots (Single Variable)$

drug.hist(column='crack-use',bins=5)   #same result

drug['crack-use'].hist(bins=5)         #same result

drug['crack-use'].plot(kind='hist',bins=5,title='title on top')   #same result

drinks.hist(column='beer', by='continent')    # bar charts grouped by continent

* $Box plots$

df[['col1']].boxplot();                # box plot of single variable

df.boxplot(column=’col1’);             # box plot of single variable

df[['col1']].plot(kind='box')          # box plot of single variable

#### Faceted lineplot using Relplot

Source: https://seaborn.pydata.org/examples/faceted_lineplot.html

```python
sns.set(style="ticks")

dots = sns.load_dataset("dots")

# Define a palette to ensure that colors will be
# shared across the facets
palette = dict(zip(dots.coherence.unique(),
                   sns.color_palette("rocket_r", 6)))

# Plot the lines on two facets
sns.relplot(x="time", y="firing_rate",
            hue="coherence", size="choice", col="align",
            size_order=["T1", "T2"], palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots)
```

##### Discovering structure in a heatmap using Cluster map

Source: https://seaborn.pydata.org/examples/structured_heatmap.html

```python
sns.set()

# Load the brain networks example dataset
df4 = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

# Select a subset of the networks
used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df4.columns.get_level_values("network")
                          .astype(int)
                          .isin(used_networks))
```

```python
df4 = df4.loc[:, used_columns]

# Create a categorical palette to identify the networks
network_pal = sns.husl_palette(8, s=.45)
network_lut = dict(zip(map(str, used_networks), network_pal))

# Convert the palette to vectors that will be drawn on the side of the matrix
networks = df4.columns.get_level_values("network")
network_colors = pd.Series(networks, index=df4.columns).map(network_lut)

# Draw the full plot
sns.clustermap(df4.corr(), center=0, cmap="vlag",
               row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13))
```

### 22) Correlation matrix

Adapted from: https://seaborn.pydata.org/examples/many_pairwise_correlations.html

```python
# Compute the correlation matrix
corr = df4.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
